## 1. Einführung & Datensatz

### 1.1 Datenquelle & Beschreibung

**Datensatz:**  
Der vorliegende Datensatz stammt von Kaggle und enthält transaktionale Verkaufsdaten eines Online-Einzelhändlers. Er dokumentiert Kundenbestellungen mit demografischen Informationen und Produktdetails.

**Quelle:**
- **Plattform:** Kaggle - Online Shop Customer Sales Data
- **Link:** https://www.kaggle.com/datasets/onlineretailshop/online-shop-customer-sales-data
- **Art:** Transaktionsdaten eines internationalen Online-Shops
- **Umfang:** Historische Verkaufsdaten mit Kunden- und Produktinformationen

**Zweck der Analyse:**  
Dieser Datensatz ermöglicht die Analyse von Kaufverhalten, Umsatzmustern und Kundensegmenten, um wertvolle Erkenntnisse für strategische Geschäftsentscheidungen zu gewinnen.

### 1.2 Variablen im Datensatz

Der Datensatz umfasst folgende Spalten:

| Variable | Datentyp | Beschreibung |
|----------|----------|--------------|
| **Transaction ID** | Numerisch | Eindeutige Identifikationsnummer für jede Transaktion |
| **Date** | Datum | Zeitpunkt der Transaktion |
| **Customer ID** | Numerisch | Eindeutige Kundenidentifikation |
| **Gender** | Kategorial | Geschlecht des Kunden (Male/Female) |
| **Age** | Numerisch | Alter des Kunden in Jahren |
| **Product Category** | Kategorial | Produktkategorie (z.B. Electronics, Clothing, Beauty) |
| **Quantity** | Numerisch | Anzahl der gekauften Produkte pro Transaktion |
| **Price per Unit** | Numerisch | Einzelpreis pro Produkteinheit |
| **Total Amount** | Numerisch | Gesamtumsatz der Transaktion (Quantity × Price per Unit) |

### 1.3 Potenzielles Businessproblem

**Ausgangslage:**  
Viele Online-Retailer stehen vor der Herausforderung, ihre Ressourcen optimal einzusetzen und datenbasierte Entscheidungen zu treffen. Ohne systematische Analyse bleiben wichtige Potenziale oft ungenutzt.

**Potenzielle Problemfelder:**

**1. Ineffiziente Ressourcenallokation:**
- Marketing-Budget wird möglicherweise nicht optimal auf Produktkategorien und Zielgruppen verteilt
- Unklare Priorisierung zwischen verschiedenen Kundensegmenten
- Potenzial für Umsatzsteigerung durch bessere Fokussierung

**2. Fehlende Transparenz:**
- Welche Produktkategorien haben das größte Wachstumspotenzial?
- Welche Kundengruppen sind am wertvollsten?
- Gibt es saisonale Muster, die für Kampagnenplanung genutzt werden könnten?

**3. Strategische Unsicherheit:**
- Entscheidungen basieren oft auf Intuition statt auf Daten
- Cross-Selling- und Upselling-Potenziale werden nicht systematisch genutzt
- Kundenbindung könnte durch personalisierte Ansprache verbessert werden

**Relevanz für die Praxis:**

Die Analyse dieses Datensatzes kann helfen:
- **Umsatzpotenziale** zu identifizieren und zu nutzen
- **Marketing-Effizienz** durch Zielgruppen-Fokussierung zu steigern
- **Kundenzufriedenheit** durch personalisierte Angebote zu erhöhen
- Eine **datenbasierte Entscheidungsgrundlage** für Sortiment, Pricing und Kampagnen zu schaffen

### 1.4 Analytische Fragestellungen

Um die potenziellen Geschäftsprobleme zu adressieren, werden folgende Fragen untersucht:

**Umsatz & Performance:**

1. Wie entwickelt sich der Umsatz über den Zeitverlauf? Gibt es Trends oder saisonale Muster?
2. Welche Produktkategorien generieren den höchsten Umsatz und die meisten Transaktionen?

**Kundensegmentierung:**

3. Welche Altersgruppen sind die wertvollsten Kunden (höchster durchschnittlicher Bestellwert)?
4. Gibt es signifikante Unterschiede im Kaufverhalten zwischen männlichen und weiblichen Kunden?

**Strategische Insights:**

5. Welche Produkt-Kundensegment-Kombinationen sind besonders profitabel?
6. Wo liegen ungenutzte Umsatzpotenziale für Cross-Selling und Upselling?
